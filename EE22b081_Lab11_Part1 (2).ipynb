(a ) Plot the sigmoid function. Print your interpretation on why this function is useful for a classification problem.

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = sigmoid(x) #function to be plotted

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sigmoid Function', color='purple')
plt.title('SIGMOID FUNCTION')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()
plt.show()
```

(b) Plot the log functions in the cost function individually. Print your interpretation of the log functions

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Cost function for the positive class
def cost_positive(y_pred):
    return -np.log(y_pred)

# Cost function for the negative class
def cost_negative(y_pred):
    return -np.log(1 - y_pred)

x = np.linspace(0.01, 0.99, 100) 

y_positive = cost_positive(x)
y_negative = cost_negative(x)

# Plotting the positive class cost function
plt.figure(figsize=(8, 6))
plt.plot(x, y_positive, label='Positive Class', color='green')
plt.title('COST FUNCTION FOR POSITIVE CLASS')
plt.xlabel('Predicted Probability')
plt.ylabel('Cost')
plt.grid(True)
plt.legend()
plt.show()

# Plotting the negative class cost function
plt.figure(figsize=(8, 6))
plt.plot(x, y_negative, label='Negative Class', color='red')
plt.title('COST FUNCTION FOR NEGATIVE CLASS')
plt.xlabel('Predicted Probability')
plt.ylabel('Cost')
plt.grid(True)
plt.legend()
plt.show()

print("1.The output of the sigmoid function is bounded between 0 and 1. This property is particularly useful in binary classification problems where we want to assign probabilities to the class labels. The output can be interpreted as the probability of belonging to a particular class.")
print("2. It interprets the model's predictions in terms of probabilities.")
print("3.The sigmoid function is smooth and continuously differentiable across its domain. This property facilitates the learning process in optimization algorithms like gradient descent, making it more stable and efficient.")
```

c) Using your own data for a single feature problem, and assuming linear regression problem, plot the cost function and the corresponding contours. Also, using cross entropy as the cost function, plot it as well as its contours.

```python
import numpy as np
import matplotlib.pyplot as plt

# Linear Regression Cost Function
def linear_regression_cost(theta0, theta1, x, y):
    cost = np.mean((theta0 + theta1*x - y)**2)
    return cost

# Cross Entropy Cost Function
def cross_entropy_cost(theta0, theta1, x, y):
    epsilon = 1e-10
    cost = -np.mean(y * np.log(theta0 + theta1*x + epsilon) + (1 - y) * np.log(1 - (theta0 + theta1*x) + epsilon))
    return cost

# Generate sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])  

# Create meshgrid for contour plot
theta0_vals = np.linspace(-3, 3, 100)
theta1_vals = np.linspace(-3, 3, 100)
T0, T1 = np.meshgrid(theta0_vals, theta1_vals)

# Calculate costs for each combination of theta0 and theta1
Z_linear = np.array([linear_regression_cost(t0, t1, x, y) for t0, t1 in zip(np.ravel(T0), np.ravel(T1))])
Z_linear = Z_linear.reshape(T0.shape)

Z_cross_entropy = np.array([cross_entropy_cost(t0, t1, x, y) for t0, t1 in zip(np.ravel(T0), np.ravel(T1))])
Z_cross_entropy = Z_cross_entropy.reshape(T0.shape)

# Plot Linear Regression Cost Function
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.contour(T0, T1, Z_linear, levels=20)
plt.xlabel('Theta0')
plt.ylabel('Theta1')
plt.title('Linear Regression Cost Function Contour Plot')

# Plot Cross Entropy Cost Function
plt.subplot(1, 2, 2)
plt.contour(T0, T1, Z_cross_entropy, levels=20)
plt.xlabel('Theta0')
plt.ylabel('Theta1')
plt.title('Cross Entropy Cost Function Contour Plot')

plt.tight_layout()
plt.show()
```
