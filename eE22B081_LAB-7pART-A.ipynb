```python
#Q1. Write a program that takes coefficients A, B, C, D, and E as inputs representing a 4th degree
#polynomial in the form Ax^4 + Bx^3 + Cx^2 + Dx + E. Calculate the values of this polynomial for 
#x in the range from -100 to 100, with constant discrete intervals. 

import numpy as np
import matplotlib.pyplot as plt

def polynomial_func(x, A, B, C, D, E):
    return A * x**4 + B * x**3 + C * x**2 + D * x + E

#INPUT COEFFICIENTS
A = 1
B = 2
C = 3
D = 4
E = 5

#GENERATING X AND Y VALUES
x = np.arange(-100, 101, 1)
y = polynomial_func(x, A, B, C, D, E)

data = np.column_stack((x, y))

#GRAPH PLOT
plt.plot(data[:, 0], data[:, 1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('4th Degree Polynomial')
plt.show()
```

```python
# Q2. Write a Python script to convert this dictionary into a pandas DataFrame,
#calculate the total sales for each product over the entire period, and then create a bar plot 
#using matplotlib to visualize the total sales for each product. 

import pandas as pd
import matplotlib.pyplot as plt

sales_data = {

    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],

    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],

    'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]

}

#CREATING A DATAFRAME FROM DICTIONARY
sales_df = pd.DataFrame(sales_data)

#CALCULATION OF TOTAL SALES
total_sales = sales_df.groupby('Product').sum()

#BAR GRAPH
plt.bar(total_sales.index, total_sales['Sales'])
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Total Sales for Each Product')
plt.show()
```

```python
#Q3. Create visualizations for the following mathematical functions using Matplotlib:

import numpy as np
import matplotlib.pyplot as plt

# Single-variable functions
x = np.linspace(-10, 10, 1000)

# 1. y = cos(x)
y_cos = np.cos(x)
plt.figure()
plt.plot(x, y_cos)
plt.title('Plot of y = cos(x)')
plt.xlabel('x')
plt.ylabel('y')

# 2. y = e^x
y_exp = np.exp(x)
plt.figure()
plt.plot(x, y_exp)
plt.title('Plot of y = e^x')
plt.xlabel('x')
plt.ylabel('y')

# 3. y = log(x), where x>0
x_log = np.linspace(0.01, 10, 1000)
y_log = np.log(x_log)
plt.figure()
plt.plot(x_log, y_log)
plt.title('Plot of y = log(x), where x > 0')
plt.xlabel('x')
plt.ylabel('y')

# MULTI-VARIABLE FUNCTIONS
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# GENERATING SURFACE PLOTS FOR MULTI-VARIABLE FUNCTIONS 

# 1. z = cos(sqrt(x^2+y^2))
Z_cos = np.cos(np.sqrt(X**2 + Y**2))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z_cos)
ax.set_title('Surface Plot of z = cos(sqrt(x^2+y^2))')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# 2. z = e^(-(x^2+y^2))
Z_exp = np.exp(-(X**2 + Y**2))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z_exp)
ax.set_title('Surface Plot of z = e^(-(x^2+y^2))')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# 3. z = log(x^2+y^2) where x^2+y^2>0
Z_log = np.log(X**2 + Y**2)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z_log)
ax.set_title('Surface Plot of z = log(x^2+y^2) where x^2+y^2>0')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

```python
#Q4. For the function J(w) = w^2 + (54/w), implement the bracketing method

import numpy as np
# Defining the function J(w)
def J(w):
    return w**2 + 54/w

# using the Bracketing method 
def bracketing_method(a, b, n):
    for i in range(n):
        c = (a + b) / 2
        if J(a) * J(c) < 0:
            b = c
        elif J(b) * J(c) < 0:
            a = c
        else:
            break
    return c

# Initial interval [a, b] and number of iterations
a = 1
b = 5
n = 100

root = bracketing_method(a, b, n)

print(f"The root of the function is approximately: {root}")
```
